```python
#Generate random fields
from numpy import random

N=100
#Create Uniform Distribution
x1 = 2*random.rand(N)-1  # Uniform distribution centered at 0, varying between -1 and 2
#Gaussian Distribution
x2 = random.randn(N) #G Gaussian distribution centered at 0
#Power Law
a=2 # power of the distribution
x3 = random.power(a,N)
# Poisson
aa=4
x4 = random.poisson(aa,N)
```

```python
fig1, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize = (10, 4))
ax1.plot(x1)
ax2.plot(x2)
ax3.plot(x3)
ax4.plot(x4)

fig2, (ax11, ax12, ax13, ax14) = plt.subplots(1,4, figsize = (10,4))
ax11.hist(x1)
ax12.hist(x2)
ax13.hist(x3)
ax14.hist(x4)

plt.show()
```
